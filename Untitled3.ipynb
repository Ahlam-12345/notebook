```python
# Import Sections
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from tensorflow.keras.datasets import mnist
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import seaborn as sns

from time import time
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load the data from keras
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
# Data exploration
print(f"Training set: {x_train.shape}")
print(f"Target in training set: {y_train.shape}")
print(f"Test set: {x_test.shape}")
print(f"Target in test set: {y_test.shape}")
```

```python
# Display the uniqe labels and their counts
unique_labels, label_counts = np.unique(y_train, return_counts=True)
print(f"Unique labels: {unique_labels}")
print(f"Number of samples per label")
for label, count in zip(unique_labels, label_counts):
    print(f"Label {label}: {count}")
```

```python
# Visualize the orginal digits
plt.figure(figsize=(10, 10))
for i in range(5):
  plt.subplot(1,5, i+1)
  plt.imshow(x_train[i], cmap='gray')
  plt.axis('off')
  plt.title(f"Label: {y_train[i]}")
plt.show()
```

```python
# Flatten the images and combine train/test data
x_train_flat = x_train.reshape(x_train.shape[0], -1)
x_test_flat = x_test.reshape(x_test.shape[0], -1)
x = np.concatenate([x_train_flat, x_test_flat])
y = np.concatenate([y_train, y_test])
```

```python
x.shape
```

```python
# Take the sample for faster processing **
n_samples = 10000
x = x[:n_samples]
y = y[:n_samples]
```

```python
x.shape
```

```python
# Split the data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
# Standadize the data (feature scaling)
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

Implementing PCA


```python
# First PCA to analyze the explained variance ratio with all compnents
pca_all = PCA()
pca_all.fit(x_train)
```

```python
# Plot the explained variance
plt.figure(figsize=(10, 5))
cumsum = np.cumsum(pca_all.explained_variance_ratio_)
plt.plot(cumsum)
plt.axhline(y = 0.9,color = 'r', linestyle = "--",label = '90% variance explained')
plt.xlabel('Number of components')
plt.ylabel('Cumulative explained variance')
plt.grid(True)
plt.legend()
plt.show()

# Find the number of components needed for 90% variance explained
n_components = np.argmax(cumsum >= 0.9) + 1
print(f"Number of components needed for 90% variance explained: {n_components}")
```

```python
# PCA with 50 components
pca_50 = PCA(n_components = 50)
x_train_pca = pca_50.fit_transform(x_train)
x_test_pca = pca_50.transform(x_test)
```

```python
# Reconstructing the image of digits with 50 components
x_train_pca_recon = pca_50.inverse_transform(x_train_pca)

plt.figure(figsize=(10, 10))
for i in range(5):
    plt.subplot(1, 5, i+1)
    plt.imshow(x_train_pca_recon[i].reshape(28,28), cmap='gray')
    plt.axis('off')
    plt.title(f"Label: {y_train[i]}")

plt.show()
```

Implementing LDA

```python
# Applyig LDA with 9 components
lda = LDA(n_components=9)
x_train_lda = lda.fit_transform(x_train, y_train)
x_test_lda = lda.transform(x_test)
```

```python
# Visualize the 2D data - PCA
plt.figure(figsize=(10, 5))
scatter = plt.scatter(x_train_pca[:, 0], x_train_pca[:, 1], c=y_train, cmap='rainbow')
plt.xlabel("First Principal Component")
plt.ylabel("Second Principal Component")
plt.legend(*scatter.legend_elements(), title="Digits")
plt.show()
```

```python
# Visualize the 2D data - LDA
plt.figure(figsize=(10, 5))
scatter = plt.scatter(x_train_lda[:, 0], x_train_lda[:, 1], c=y_train, cmap='rainbow')
plt.xlabel("First Principal Component")
plt.ylabel("Second Principal Component")
plt.legend(*scatter.legend_elements(), title="Digits")
plt.show()
```

Classification And Evaluation

```python
from sklearn.neighbors import KNeighborsClassifier

print(f"Training and evaluating the KNN classifier with ORIGINAL Data")
knn_original = KNeighborsClassifier(n_neighbors=5)
knn_original.fit(x_train, y_train)
y_pred_original = knn_original.predict(x_test)
accuracy_original = accuracy_score(y_test, y_pred_original)
print("-"*50)

print(f"Training and evaluating the KNN classifier with PCA Data")
knn_pca = KNeighborsClassifier(n_neighbors=5)
knn_pca.fit(x_train_pca, y_train)
y_pred_pca = knn_pca.predict(x_test_pca)
accuracy_pca = accuracy_score(y_test, y_pred_pca)
print("-"*50)

print(f"Training and evaluating the KNN classifier with LDA Data")
knn_lda = KNeighborsClassifier(n_neighbors=5)
knn_lda.fit(x_train_lda, y_train)
y_pred_lda = knn_lda.predict(x_test_lda)
accuracy_lda = accuracy_score(y_test, y_pred_lda)
print("-"*50)
print('PerformanceResults:')
print(f"Original data:{x_train.shape[1]} dimensions,Accuracy : {accuracy_original}")
print(f"PCA data:{x_train_pca.shape[1]} dimensions,Accuracy : {accuracy_pca}")
print(f"LDA data:{x_train_lda.shape[1]} dimensions,Accuracy : {accuracy_lda}")

print("-"*50)
print('Memory Usage:')
print(f"Original data: {x_train.nbytes / (1024**2):.2f} MB")
print(f"PCA data: {x_train_pca.nbytes / (1024**2):.2f} MB")
print(f"LDA data: {x_train_lda.nbytes / (1024**2):.2f} MB")
```
